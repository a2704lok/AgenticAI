````python
from PyPDF2 import PdfReader

def read_pdf(file_path):
    reader = PdfReader(file_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"
    return text.strip()
````

````python
# Using another model to evaluate a model
````

````python
model1 = ChatGroq(model = "Gemma2-9b-It",groq_api_key = groq_api_key)
model2 = ChatGroq(model = "llama3-70b-8192",groq_api_key = groq_api_key)
````

````python
def generate_response(input_prompt: str) -> str:
    messages = [
        SystemMessage(content="You are a helpful assistant."),
        HumanMessage(content=input_prompt)
    ]
    response = model1.invoke(messages)
    return parser.invoke(response)

# Function to evaluate model1's output using model2
def evaluate_response(input_prompt: str, response: str) -> tuple[bool, str]:
    eval_prompt = (
        f"The following response was generated for the prompt:\n\n"
        f"PROMPT:\n{input_prompt}\n\nRESPONSE:\n{response}\n\n"
        f"Is this response accurate, helpful, and relevant? "
        f"Reply with 'Yes' if acceptable, otherwise reply with 'No' and explain why."
    )
    eval_messages = [
        SystemMessage(content="You are a strict evaluator of assistant responses."),
        HumanMessage(content=eval_prompt)
    ]
    eval_result = model2.invoke(eval_messages)
    eval_text = parser.invoke(eval_result)

    # Determine pass/fail
    is_valid = eval_text.strip().lower().startswith("yes")
    return is_valid, eval_text

# Function to regenerate response using evaluation feedback
def regenerate_response(input_prompt: str, feedback: str) -> str:
    regen_prompt = (
        f"The previous response to this prompt was found to be inadequate due to the following reason:\n"
        f"{feedback}\n\n"
        f"Please generate a better response to the original prompt:\n{input_prompt}"
    )
    messages = [
        SystemMessage(content="You are a helpful assistant."),
        HumanMessage(content=regen_prompt)
    ]
    response = model1.invoke(messages)
    return parser.invoke(response)

# Full pipeline
def generate_with_validation(input_prompt: str) -> str:
    initial_response = generate_response(input_prompt)
    is_approved, feedback = evaluate_response(input_prompt, initial_response)

    if is_approved:
        return f"âœ… Approved Response:\n{initial_response}"
    else:
        final_response = regenerate_response(input_prompt, feedback)
        return f"ğŸ” Regenerated Response:\n{final_response}\n\nğŸ“ Feedback:\n{feedback}"

prompt = "What is the newton's second law of motion and how does it apply to everyday life?"
result = generate_with_validation(prompt)
print(result)
````
