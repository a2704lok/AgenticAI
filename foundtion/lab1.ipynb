````python
from PyPDF2 import PdfReader

def read_pdf(file_path):
    reader = PdfReader(file_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"
    return text.strip()
````

````python
reader = read_pdf("AlokKumar_ML_DataScience_Resume_2025.pdf")
````

````python
def ask_about_candidate(resume_text: str, profession: str) -> str:
    messages = [
        SystemMessage(
            content="You are acting as HR to shortlist candidates based on their resume. The resume is in plain text."
        ),
        HumanMessage(
            content=f"Can this person be hired as a {profession} based on the resume?\n\nResume:\n```{resume_text}```"
        )
    ]
    result = model.invoke(messages)
    return parser.invoke(result)
````

````python
ask_about_candidate(reader, "doctor")
````

````python
def ask_about_candidate(resume_text: str, profession: str, skills: str) -> str:
    messages = [
        SystemMessage(
            content="You are acting as HR to shortlist candidates based on their resume.Some key skills are required for a particular role that the candidate must have. The resume is in plain text."
        ),
        HumanMessage(
            content=f"Can this person be hired as a {profession} based on the resume? Candidate muskt have skills - {skills} \
            Answer in Yes or no only. ```{resume_text}```"
        )
    ]
    result = model.invoke(messages)
    return parser.invoke(result)
````

````python
print(ask_about_candidate(reader,"data analyst", "aws, python, sql, tableau, excel"))
print(ask_about_candidate(reader,"lawyer", "notary"))
print(ask_about_candidate(reader,"machine learning engineer", "terraform, gcp"))
print(ask_about_candidate(reader,"machine learning engineer", "sagemaker, gcp"))
print(ask_about_candidate(reader,"machine learning engineer", "snowflake, gcp, airflow, kubernates"))
print(ask_about_candidate(reader,"doctor", "aws, python, sql, tableau, excel"))
````
